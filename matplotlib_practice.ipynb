```python
# Call %matplotlib inline to ensure charts are displayed within the same Jupyter Notebook cell
%matplotlib inline
# import Matplotlib's pyplot
# pyplot provides a MATLAB-like method for creating charts (simpler)

import matplotlib.pyplot as plt
```

```python
# Add starter data

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Use the plot() function from pyplot, taking the x axis as the first argument and the y axis list as the second (MATLAB INTERFACE)
plt.plot(x_axis, y_axis)
```

```python
# OOP INTERFACE PLOT METHOD
# use ax.plot()

# Create a plot using ax.plt()

# return a tuple which contains variables fig and ax (fig, ax)
# fig references figure
# ax references axes, allowing labeling

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

```

```python
# OOP -- Alternative method to above

# Create the plot with ax.plt()
# We create a figure object and assign it to fig
fig = plt.figure()

# We take the figure object fig and assign it to a subplot using the add_subplot() function
# Takes arguments as a dimension: add_subplot(111) creates one subplot with 1:1 dimensions. 

# check https://stackoverflow.com/questions/54508045/how-the-add-subplotnrows-ncols-index-works for subplot information
beep = fig.add_subplot(111)

# plots the graph with the above defined parameters, taking x and y axis data from x_axis and y_axis
beep.plot(x_axis, y_axis)
```

```python
# OOP -- Another method
# https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.axes.html
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# We can change figure attributes like labels, the title, or add a legend using the following syntax:
fig = plt.figure() 
```

```python
# Create a bar chart using the bar() function
plt.bar(x_axis,y_axis)
```

```python
# Annotating the chart (MATLAB METHOD) -----------------

# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Creating a HORIZONTAL BAR chart using the barh() function and INVERTING an axis! (MATLAB METHOD)

# Create the plot, giving it a magenta color and this series of data the label "Boston"
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
# label the x axis
plt.xlabel("Fare ($)")
# label the y axis
plt.ylabel("Date")
# give the plot a title
plt.title("PyBer Fare by Month")
# place a legend on the plot
plt.legend()
# Can swap arguments to produce opposite, but is not very valuable. We want to usually show VALUES on the X axis, not discrete categories!

# plt.barh(y_axis, x_axis)

# If we wanted to reorder the months so it's ascending from January -> December, we can use the .gca() method on the plot, as well as the .invert_yaxis() method
# the gca() method is the "get current axes" method. We can chain the .invert_yaxis() on the "get current axes" method to get the current axes of the specified plot, and then invert the y axis

plt.gca().invert_yaxis()

```

```python
# Create the plot (MATLAB METHOD)
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Creating Bar Charts using the OOP Method

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)

# The bar() function here takes the same color, label arguments as the MATLAB-based solution above
```

```python
# Creating a VERTICAL Bar Chart using the OOP Method

# Create the plot with ax.plt()
fig, ax = plt.subplots()
# color in OOP method can take cyan or CYMK 
ax.barh(x_axis, y_axis, color="y", label="Chicago")
ax.set_title("PyBer Fare by Month")
ax.legend()
# Using OOP, invert y axis
ax.invert_yaxis()
# set a label for the x axis
ax.set_xlabel("Fare ($)")

```

```python
# Scatter plots! (MATLAB METHOD) ------------------------

# Can use plt.plot() or plt.scatter()
# Recall that .plot() made LINE GRAPHS by default. Passing the argument "o" as a parameter removes the line and leaves the marker, making a scatter plot



plt.plot(x_axis, y_axis, 'o')


```

```python
# .scatter() does the same thing
plt.scatter(x_axis, y_axis, c="red", label="Chicago")
plt.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Month")
plt.ylabel("Fare ($)")
```

```python
# Bubble Plots (MATLAB METHOD) ------------------------

# We want to be able to display data sometimes in multiple dimensions. Bubble plots respect the relationship of a scatter plot but expand it into a third dimension:
# The SIZE of the bubble! We can plot x axis, y axis, size of bubble, and even COLOR of bubble to get a FOURTH DIMENSION

# the s argument changes the relative size of the data points based off the given parameter's value. In this case, the higher up the value is on the y axis, the bigger the bubble.
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# If we wanted to make these bubbles bigger, we can multiply them by a common factor:

# Here we initiate an empty list, loop through the original y axis data, and multiply each value by a factor of 3
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

# We can then modify what the s= parameter is set to, and set it to our newly created list y_axis_larger
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# LIST COMPREHENSIONS!!!!!!!!! ------------------------------------------------------

# we can refactor and optimize for loops using LIST COMPREHENSIONS. General syntax is:

# new_list = [expression for item in list if conditional]

# The following code is the same as separately for looping through our original y axis and creating a new list
# In the following:
    # expression = i * 3
    # item = i
    # list = y_axis
    # NO CONDITIONAL
    
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Scatter Plots using OOP METHOD! -----------------------------

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Changing the size of bubbles in a bubble chart (OOP METHOD ----)

fig, ax = plt.subplots()
# passing in the same argument, with s= the size scalar list
# use a LIST comprehension for the size argument instead for cleaner code
# for every item i in list y_axis, multiply i by 5
# linewidths argument is the outline weight
# edgecolors defines the color of bubbles' edge
ax.scatter(x_axis, y_axis, s=[i * 5 for i in y_axis], c="skyblue", label="Boston",
            linewidths=2, edgecolors="black")
# set limits for x axis data (months). Note a string is passed in instead of numbers as this data is ordinal
ax.set_xlim("Jan","November")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Fare ($)")

# Alternatively, we can switch axes as well by swapping in the order of arguments into the .scatter() function
```

```python
# Pie Charts! - MATLAB APPROACH ---------------
# create a pie chart using .pie()
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Bring attention to one data element from the pie using the explode argument
# Note that we have to pass in an equal-length list (or tuple list) with values pertaining to how "much" the slices are exploded by

# autopct argument adds percentage labels to each slice, using percentage formatting
    # Percentage formatting is the following syntax: "%<your value>%"
    # Respects string formatting rules too. ".1f%" displays one decimal as float and adds the % sign afterwards
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Increase the size of the chart by rescaling the subplot. Call this BEFORE PLOTTING:
plt.subplots(figsize=(8, 8))

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')

```

```python
# Note based off above, you can specify "color cycles", or premade color schemes, to Matplotlib. Without specifying, it will randomly choose within the current "color cycle"

# Matplotlib documentation for color names: https://matplotlib.org/3.1.1/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Pie Charts with the OOP Approach --------------------------------

explode_values = (0, 0, 0, 0.2, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Increase graph size
fig, ax = plt.subplots(figsize=(8, 8))



# Using matplotlib.pyplot.pie documentation, find the following arguments
# startangle sets the starting angle of the pie chart. We use 90 to rotate the graph by 90 degrees. Can take negative degrees too
# counterclock changes the order of values, either clockwise or counterclockwise against a Boolean
# autopct allows for percent formatting
ax.pie(y_axis, labels=x_axis, 
                autopct="%.1f%%", 
                explode=explode_values, 
                shadow=True,
                startangle=90,
                counterclock=False)
plt.show()
```

```python
# Import dependencies
# statistics is built into Anaconda and can be used to calculate stdev
import statistics

# Sample data in list form:

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# yerr provides the amount of error across the Y axis (calculated stdev from above)
# capsize adds caps to the error bars to look more professional
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)

# Interpretation: the extent of the bar is 1 standard deviation above and below the value. Since, when drawing a horizontal line across all values in the series, the line intersects multiple values, we know the standard deviations overlap and there is little overall variation in the values
```

```python
# Error bars and capsizes, OOP Approach -------------------------------------------

fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Changing Major Axis ticks! --------------

# Matplotlib will automatically determine a tick interval based off the range of data provided in a given plot
# Modifying this can help with readability

# use the xticks() function in conjunction with Numpy!
    # here we use Numpy to quickly create a 1 dimensional list of values from 0 to 50 (excluding last number), in increments of five (one line of code!)
    # np.arange(0, 51, step=5.0) returns a list: [ 0.  5. 10. 15. 20. 25. 30. 35. 40. 45. 50.]
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# # Changing Major Axis ticks! -------------- (OOP Method)

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()

# Alternatively, you can set the axis ticks using: ax.xaxis.set_ticks()
# this first gets the xaxis class of the ax object and calls the generic set_ticks() function on it
```

```python
# Setting Minor Ticks -------------------

from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
# We use the .set_minor_locator and MultipleLocator() method from matplotlib

# Documentation for multiple locator: https://matplotlib.org/3.1.0/gallery/ticks_and_spines/major_minor_demo.html

ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
